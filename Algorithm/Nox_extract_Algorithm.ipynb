```python
%matplotlib inline
import pandas as pd                #  data store and treat package pandas 
import matplotlib as mpl           # graph package
import matplotlib.pyplot as plt    # graph package
import numpy as np
```

```python
Nox_information = pd.read_csv('K_SW_NOX.csv',encoding='euc-kr')
Nox_information
```

현재 Nox csv 파일에 있는 값들은 선형분석으로 추출된 값이며, No2가 아닌 Nox이다. 또한 EPA 제공하고 있는 값들은 ppb 단위 이지만, 현재 이 데이터에는 ppm 단위이다. 따라서 이 Nox 선형값을 no2 로 바꾸고 단위를 ppb 단위로 변경해야 한다. 

### No2 = 97 % NO + 3 % No2 
### Nox(ppb) = Nox(ppm) x 0.3 x 1000

```python
Nox_information['NOX']
```

```python
data = Nox_information['NOX']
```

```python
data[1] # sample_testing 
```

```python
dic = {'1일차' : [], '2일차' : [], '3일차' : [], '4일차' : [], '5일차' : [], }
print(dic)
# 5일 동안 120개 데이터
```

```python
count = 0
num = 0
day = 1
while day < 6:
    if num == 120:
        day += 1 
        num = 0
    else:    
        dic["{}일차".format(day)].append(data[count])
        count += 1
        num += 1
```

```python
# print(dic['1일차'])
# 1일차 평균 Nox(ppb) = Nox(ppm) x 0.3 x 1000
average_1 = ( np.mean(dic['1일차']) ) * (0.03) * (1000)
print('1일차 평균',average_1)
# 2일차 평균
average_2 = np.mean(dic['2일차']) * (0.03) * (1000)
print('2일차 평균',average_2)
# 3일차 평균
average_3 = np.mean(dic['3일차']) * (0.03) * (1000)
print('3일차 평균', average_3)
# 4일차 평균
average_4 = np.mean(dic['4일차']) * (0.03) * (1000)
print('4일차 평균', average_4)
# 5일차 평균
average_5 = np.mean(dic['5일차']) * (0.03) * (1000)
print('5일차 평균', average_5)
Measurement_no2 = [average_1,average_2,average_3,average_4,average_5]
```
